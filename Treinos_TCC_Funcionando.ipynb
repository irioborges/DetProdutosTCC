#Treino da rede neural através do YOLOv3

```
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="/content/drive/My Drive/DataSetTCC/modelPre-classificador/")
trainer.setTrainConfig(object_names_array=["produto"], batch_size=4, num_experiments=200, train_from_pretrained_model="/content/drive/My Drive/DataSetTCC/modelPre-classificador/pretrained-yolov3.h5")
trainer.trainModel()
```
